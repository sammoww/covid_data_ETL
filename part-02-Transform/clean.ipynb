```python
#function to clean dataset

def clean(df):
    clean_df = df.drop(columns=['Province/State','Lat','Long']).\
                                                iloc[:,np.r_[0:1,40:67]].\
                                                groupby('Country/Region').sum().\
                                                reset_index()
    return clean_df                                             
```

```python
#Clean the 'confirmed_data_df' dataset
confirmed_clean_df = clean(confirmed_data_df)
confirmed_clean_df
```

```python
#Clean the 'recovered_data_df' dataset
recovered_clean_df = clean(recovered_data_df)
recovered_clean_df
```

```python
#Clean the 'deaths_data_df' dataset
deaths_clean_df = clean(deaths_data_df)
deaths_clean_df.head()
```

```python
def convert(clean_df):
    converted = pd.melt(clean_df, id_vars=["Country/Region"], var_name="Date",value_name="Number")

    #Convert data type from string to datetime64 for data in column "Date"
    converted['Date'] = pd.to_datetime(converted['Date'])

    #Sort dataset by country/region and then date
    converted_df = converted.sort_values(by=["Country/Region","Date"])

    return converted_df
```

```python
confirmed_convert_df = convert(confirmed_clean_df)
recovered_convert_df = convert(recovered_clean_df)
deaths_convert_df = convert(deaths_clean_df)

deaths_convert_df
```

```python
#Create a function to calculate daily increase

def dailychange(convert_df):

    convert_df = convert_df.set_index(['Country/Region','Date'])    

    convert_df['amount_of_increase'] = convert_df.diff()

    convert_df = convert_df.reset_index()

    for i in range(1,convert_df.index.size):
        if convert_df['Country/Region'].iloc[i] != convert_df['Country/Region'].iloc[i-1]:
            convert_df.at[i, 'amount_of_increase'] = 0

    convert_df['amount_of_increase'].fillna(0, inplace=True)

    return convert_df
```

```python
#Apply function

confirmed_daily_df = dailychange(confirmed_convert_df)
recovered_daily_df = dailychange(recovered_convert_df)
deaths_daily_df = dailychange(deaths_convert_df)
```

```python
#rename columns
confirmed_daily_df.columns=['country','date','accumulate_confirmed','increase_of_confirmed']
recovered_daily_df.columns=['country','date','accumulated_recovered','increase_in_recovered']
deaths_daily_df.columns=['country','date','accumulated_deaths','increase_in_deaths']
```

```python
confirmed_daily_df["increase_of_confirmed"] = confirmed_daily_df["increase_of_confirmed"].astype(int)
recovered_daily_df["increase_in_recovered"] = recovered_daily_df["increase_in_recovered"].astype(int)
deaths_daily_df["increase_in_deaths"] = deaths_daily_df["increase_in_deaths"].astype(int)
```
